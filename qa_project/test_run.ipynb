```python
import pandas as pd

conditions = {
    "Closed-Book":       "results_closed_book.csv",
    "RAG Only":          "results_rag_only.csv",
    "CoT Only":          "results_cot_only.csv",
    "RAG + CoT":         "results_rag_cot.csv",
    "Self-Verify":       "results_rag_cot_selfverify.csv",
}

base = "/Users/krishna/Desktop/Research/qa_project/outputs/train/"

for label, fname in conditions.items():
    df = pd.read_csv(base + fname)
    print(f"\n{'='*50}")
    print(f"Condition: {label}")
    print(f"{'='*50}")
    
    # Per-dataset EM and F1
    summary = df.groupby("source_dataset")[["em","f1"]].mean().mul(100).round(1)
    summary.columns = ["EM (%)", "F1 (%)"]
    print(summary.to_string())
    
    print(f"\nOverall  EM = {df['em'].mean()*100:.1f}%   F1 = {df['f1'].mean()*100:.1f}%")
```
